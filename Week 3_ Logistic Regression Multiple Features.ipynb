## Solving the Problem of Overfitting

### The Problem of Overfitting

Sebelum kita masuk lebih jauh mengenal overfitting. Kita harus tau apa itu Bias dan apa itu Variance.

Bias adalah perbedaan antara rata-rata prediksi model kita dengan rata-rata nilai aktual yang kita coba prediksi. Penyebab bias pada model tinggi adalah kita kurang memperhatikan training set dan menyederhanakan fungsi hipotesis. Seperti kita menggunakan features yang sedikit padahal features pada training set ada sangat banyak.

Sedangkan variance adalah informasi tentang seberapa jauh persebaran data model dalam memprediksi data aktual. Model yang memiliki variance tinggi disebabkan karena model terlalu memberikan perhatian khusus pada training set. Padahal harapan kita adalah nantinya model dapat bekerja baik saat memprediksi data yang belum pernah dilihatnya. Alhasil model yang dibentuk tidak bekerja ideal untuk data yang baru dilihat.

### Cost Function

```
function [J, grad] = costFunction(theta, X, y)
%COSTFUNCTION Menghitung fungsi cost dan gradient descent
%   J = COSTFUNCTION(theta, X, y) Menghitung fungsi cost dengan theta sebagai
%   parameter untuk regresi logistic dalam menghitung gradient dan fungsi cost

% Meng-inisialisasi / membuat variabel yang berguna
m = length(y); % number of training examples

% Kamu harus mengembalikan nilai variabel J dan grad dengan benar! 
J = 0;
grad = zeros(size(theta));

% ====================== KODEMU DISINI ======================
% Instruksi: Kamu harus membuat isi dari variabel J sebagai fungsi cost
%            Hitunglah parsial derivatif dan isi variabel sebagai hasil perhitungan
%               parsial derivatif untuk setiap nilai theta
%
% Note: variabel grad harus memiliki dimensi yang sama dengan ukuran parameter theta
%

% Menghitung hipotesis
z = X * theta;
sig = sigmoid(z);

% Menghitung fungsi cost
J = J + (1 / m) * (-y' * log(sig) - (1 - y)' * log(1 - sig));

# Menghitung parsial derivatif
grad = grad + (1 / m) * (X' * (sig - y));
% =============================================================

end
```

Penjelasan Tentang Kode: Sebelum menghitung fungsi cost dan parsial derivatifnya. Saya menyiapkan variabel yang penting untuk perhitungan nanti, seperti fungsi hipotesis.

Setelah itu, saya menghitung fungsi cost dan menyimpannya dalam variabel J. Seperti yang diminta. Lihatlah rumus sederhana fungsi cost [disini](https://ml-cheatsheet.readthedocs.io/en/latest/_images/logistic_cost_function_vectorized.png).

Perhitungan fungsi parsial dari regresi logistic sama saja dengan regresi linear. Tetapi, yang perlu kalian lihat adalah fungsi hipotesisnya, itu berbeda.

### Cost Function dengan Regularization

```
function [J, grad] = costFunctionReg(theta, X, y, lambda)
%COSTFUNCTIONREG Menghitung fungsi cost dan gradient dengan Regularization
%   J = COSTFUNCTIONREG(theta, X, y, lambda) menghitung fungsi cost dengan
%   theta sebagai parameter untuk regularized regresi logistic dan 
%   juga gradient. 

% Meng-inisialisasi / membuat variabel yang berguna
m = length(y); % number of training examples

% Kamu harus mengembalikan nilai dari variabel J dan grad dengan benar! 
J = 0;
grad = zeros(size(theta));

% ====================== KODEMU DISINI ======================
% Instruksi: Kamu harus membuat isi dari variabel J sebagai fungsi cost
%            Hitunglah parsial derivatif dan isi variabel sebagai hasil perhitungan
%               parsial derivatif untuk setiap nilai theta
%
% Note: variabel grad harus memiliki dimensi yang sama dengan ukuran parameter theta
%

% Menghitung Hypotesis
z = X * theta;
sig = sigmoid(z);

% Menyiapkan costFunction dan grad tanpa di-regularized
[J, grad] = costFunction(theta, X, y);

% Menambahkan parameter penting untuk perhitungan regularized logistic function
J = J + (lambda / (2 * m)) * sum(theta(2:end) .^ 2);
grad(2:end) = grad(2:end) + (lambda / m) * theta(2:end);
% =============================================================

end
```

Penjelasan Tentang Kode: Disini saya tetap menyiapkan variabel J dan grad untuk menyimpan fungsi cost yang belum di regularized.

Karena, saya melihat rumus perhitungannya itu masih adopsi dari fungsi cost dan parsial derivatif yang biasa. Sehingga biar tidak memperbanyak penulisan kode, saya menyingkatkannya dengan memanggil fungsi costFunction yang lama.
