### Implementasi **Normal Equation**

````
import matplotlib.pyplot as plt
import numpy as np

"""
Dummy Data for Linear Regression
"""
data = [(1, 1), (2, 2), (3, 4), (4, 3), (5, 5.5), (6, 8), (7, 6), (8, 8.4), (9, 10), (5, 4)]    

"""
Matrix Operations
"""
inv = np.linalg.inv # Function for get invers of matriks
mul = np.matmul # Function for product matriks
````

Seperti yang saya beritahu sebelumnya. Kita perlu mengubah data dummy diatas menjadi 2 matriks *X* dan *y*. Tambahkan juga nilai 1 pada kolom pertama di matriks *X*. Jangan lupa men-transpose matriks *y*, karena untuk mengalikan 2 matriks yang berbeda, kita perlu pastikan bahwa baris pada matriks *X* sama dengan banyaknya baris matriks *y*.

````
X = []
y = []
for x_i, y_i in data:
    X.append([1, x_i])
    y.append(y_i)
X = np.array(X)
y = np.atleast_2d(y).T
````

Pastikan matriks telah siap, lalu kita gunakan rumus Normal Equation. Dapet deh! nilai θ yang cocok.

````
"""
Theta Calculation Using equation (5)
"""
theta = mul(mul(inv(mul(X.T, X)), X.T), y)
theta
````

````
"""
Prediction of y using theta
"""
y_pred = np.matmul(X, theta)

"""
Plot Graph
"""
plt.scatter([i[0] for i in data], [i[1] for i in data])
plt.plot([i[0] for i in data], y_pred, 'r')
plt.title('Regression using Normal Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

Setelah ini, pasti kalian berpikir lebih baik menggunakan Normal Equation dibanding Gradient Descent. Sayangnya tidak, dalam beberapa hal juga Gradient Descent lebih baik digunakan, berikut perbandingannya:

Gradient Descent
*   Perlu menentukan nilai α (alfa)
*   Butuh banyak iterasi / perulangan
*   Bekerja lebih baik dengan jumlah data yang sangat besar.

Normal Equation
*   Tidak perlu menentukan nilai α (alfa)
*   Tidak perlu melakukan iterasi / perulangan
*   Bekerja lambat dengan jumlah data yang sangat besar.

Gradient Descent lebih baik digunakan jika banyaknya baris data sudah sampai 10^6.

### Compute Cost & Compute Cost Multi Variable / Menghitung Fungsi Cost

```
function J = computeCost(X, y, theta)
% computeCost Menghitung fungsi cost pada Regresi Linear
%   J = computeCost(X, y, theta) Menghitung fungsi cost menggunakan nilai theta sebagai
%   parameter untuk mencari tahu seberapa cocok garis yang terbentuk dengan titik Y. Berdasarkan nilai X.

% Initialize some useful values
m = length(y); % Banyaknya baris pada dataset

% Kamu harus mengembalikan nilai fungsi cost dengan tepat dan benar!
J = 0;

% ====================== Kodemu Disini ======================
% Instruksi: Menghitung seberapa jauh error dari garis yang terbentuk dalam memprediksi nilai Y. 
%            Kamu harus membuat variabel J menjadi nilai dari perhitungan fungsi cost.
h = (X * theta);
err = (h - y) .^ 2;
J = sum(err) / (2 * m);
% =========================================================================

end

```

### Gradient Descent & Gradient Descent Multi Variable

```
function [theta, J_history] = gradientDescent(X, y, theta, alpha, num_iters)
% GradientDescent Melakukan gradient descent untuk belajar mencari nilai theta dengan iterasi / berulang-ulang
%   theta = gradientDescent(X, y, theta, alpha, num_iters) memperbarui theta dengan 
%   melakukan perulangan sebanyak nilai pada parameter num_iters

% Membuat variabel dengan nilai/value yang berguna
m = length(y); % Menyimpan banyaknya baris pada dataset
J_history = zeros(num_iters, 1);

for iter = 1:num_iters

    % ====================== Kodemu Disini ======================
    % Instruksi: Melakukan langkah gradient descent untuk memperbarui
    %               theta pada setiap perulangan. 
    %
    %
    h = X * theta;
    err = h - y;
    theta = theta - alpha * (X' * err / m);
    % ============================================================

    % Hint: Selama debugging, variabel J_history sangat berguna untuk mengetahui
    %       nilai dari fungsi cost dari iterasi k
    % Save the cost J in every iteration    
    J_history(iter) = computeCost(X, y, theta);

end;

end;
```
